```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")

#ploting the distribution of calorie counts
plt.figure(figsize=(10, 6))
sns.histplot(data['Calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calories Across Menu Items', fontsize=16)
plt.xlabel('Calories', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Boxplot for total fat, protein, and carbohydrates
plt.figure(figsize=(14, 8))
nutrients = ['Total Fat', 'Protein', 'Carbohydrates']
data_long = data.melt(id_vars=['Item'], value_vars=nutrients, var_name='Nutrient', value_name='Amount')

sns.boxplot(x='Nutrient', y='Amount', data=data_long, palette='Set2')
plt.title('Nutritional Content of Menu Items (Fat, Protein, Carbohydrates)', fontsize=16)
plt.xlabel('Nutrient', fontsize=12)
plt.ylabel('Amount (g)', fontsize=12)
plt.show()
```

```python
# Pairplot to examine relationships between key nutrients
sns.pairplot(data[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']], diag_kind='kde')
plt.suptitle('Pairplot of Calories, Total Fat, Protein, and Carbohydrates', y=1.02, fontsize=16)
plt.show()
```

```python
# Identifying menu items with the highest and lowest calorie counts
highest_calories = data.loc[data['Calories'].idxmax()]
lowest_calories = data.loc[data['Calories'].idxmin()]
```

```python
# Calculating the average nutritional content for popular menu categories (Total Fat, Protein, Carbohydrates, Calories)
popular_categories = data['Category'].value_counts().nlargest(5).index  # Identify the top 5 most common categories
average_nutrition = data[data['Category'].isin(popular_categories)].groupby('Category')[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']].mean()

average_nutrition
```
